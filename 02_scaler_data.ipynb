```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from joblib import dump
```

#### Escalar solo el dataset de entrenamiento

```python
df_data = pd.read_csv("process_dataset/train_data.csv")
df_data
```

#### FunciÃ³n para guardar escaladores y datos escalados

```python
def save_scaler(df_data, scaler_instance, type_scaler):
    scaler_instance.fit(X=df_data.values)
    data_scaled = scaler_instance.transform(X=df_data.values)
    df_scaled = pd.DataFrame(data=data_scaled, columns=df_data.columns)
    df_scaled.to_csv(f"process_dataset/train_data_scaled_{type_scaler}.csv", index=False)
    dump(scaler_instance, f"results/scaler_{type_scaler}.joblib")
```

#### Escalar los datos con Standard Scaler

```python
standard_scaler_instance = StandardScaler()
save_scaler(df_data, standard_scaler_instance, "standard")
```

#### Escalar los datos con MinMax Scaler

```python
min_max_scaler_instance = MinMaxScaler()
save_scaler(df_data, min_max_scaler_instance, "min_max")
```

#### Escalar los datos con Robust Scaler

```python
robust_scaler_instance = RobustScaler()
save_scaler(df_data, robust_scaler_instance, "robust")
```
