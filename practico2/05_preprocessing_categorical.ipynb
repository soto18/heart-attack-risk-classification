##### Codificaci√≥n con one-hot
- Trata todas las categor√≠as de manera equitativa, evitando que el modelo aprenda relaciones incorrectas entre ellas.
- Facilita la interpretaci√≥n

```python
df_data = pd.get_dummies(df_data, columns=['Sex', 'Diet', 'Continent'])
df_data[df_data.select_dtypes(include='bool').columns] = df_data.select_dtypes(include='bool').apply(lambda x: x.astype(int))
```
